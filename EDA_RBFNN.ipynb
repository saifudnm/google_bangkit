```
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```
  url = 'https://raw.githubusercontent.com/saifudnm/google_bangkit/master/data/shallot.csv'
  df = pd.read_csv(url)
```

```
df
```

```
df['date_baru']=df['Date'].str[:11]
df.groupby('Date',as_index=True).agg({'Shallot':sum}).plot(kind='line')
plt.ylabel('Price')
plt.title("Time Series Plot Shallot's Prices")
```

The Graph shows time series plot of shallot's prices, We can know that the highest price is around june because there was ied-fitri ceremony. The pattern is fluctuating followed by the event or ceremony in Indonesia. Indonesian's people like cooking many food using shallot to flavour the foods.

```
plt.figure(figsize=(12,5))
plt.title("Distribution of Shallot's Prices")
ax = sns.distplot(df["Shallot"], color = 'b')
```

The graph shows the distribution of shallot's price, we can know that the distribution is left skewness. The highest price is around 40000 rupiah.
